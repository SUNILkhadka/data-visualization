# Scatter Plots
Leverage the coordinate plane to explore relationships between variables

```python
data_df = pd.read_csv('/home/bing0/Documents/Internship/Github/data-visualization/data/insurance.csv')
```

```python
data_df.head()
```

```python
sns.scatterplot(x = data_df['bmi'],y = data_df['charges'])
plt.show()
```

```python
sns.regplot(x = data_df['bmi'],y = data_df['charges'])
plt.show()
```

```python
sns.scatterplot(x = data_df['bmi'],y = data_df['charges'],hue=data_df['smoker'])
plt.show()
```

```python
sns.lmplot(x='bmi',y='charges',hue='smoker',data=data_df)
```

```python
sns.swarmplot(data=data_df,y='charges',x='smoker')
```
