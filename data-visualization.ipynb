```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data_df = pd.read_csv('/home/bing0/Documents/Internship/Github/data-visualization/museum_visitors.csv',parse_dates=True)
```

```python
data_df['Date'] = pd.to_datetime(data_df['Date'])
```

```python
data_df.head()
```

```python
data_df.info()
```

## lineplot

```python
fig = plt.figure(figsize=(14,5))
sns.lineplot(data = data_df)
plt.show()
```

## barplot

```python
sns.barplot(x=data_df['Date'].dt.month,y=data_df['Firehouse Museum'])
plt.title('Firehouse Museum data on the basis of months')
plt.show()
```
